````python
#1) Seu código aqui
# Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de 
#    interessado ('Estado'ou Município), e que devolve os dados da consulta no formato DataFrame.

# _________________________________________________

### comparando as urls:

url_aula = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
url_exer = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Estado'

# oberseva-se que a url muda em uf=[sigla do estado], vamos analisar uma url para comparar quando selecionado um "Município" no lugar de "Estado"

url_muni = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio'

# então, quando selecionado "Municipio", a url muda para tipo_interessado=[Munic%C3%ADpio]
````

````python
# após a compação, criando a função:

def dados_pvl(UF:str , tipo_interessado:str):
    if tipo_interessado == 'Estado':
        tipo_interessado = 'Estado'
    elif tipo_interessado == 'Município':
        tipo_intessado = 'Munic%C3%ADpio'    
        
    url_pesq = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=' +UF.upper()+ '&tipo_interessado=' +tipo_interessado
    r_pesq = requests.get(url_pesq)
    
    return pd.DataFrame(r_pesq.json()['items'])

````

````python
# 2) Seu código aqui
# Quantas solicitações para o Estado podem ser 
      #consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?

consulta_mg = dados_pvl('MG', 'Estado')

consulta_mg[consulta_mg['status'] == 'Arquivado por decurso de prazo'].status.value_counts()
````

````python
# 3) Seu código aqui
# Qual é o município da Bahia com mais solicitações deferidas?

consulta_ba_2 = dados_pvl('BA','Município')
consulta_ba_2[consulta_ba['status']=='Deferido']['interessado'].value_counts().head(1)

````
